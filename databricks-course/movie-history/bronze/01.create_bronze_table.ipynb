```sql
create schema if not exists movie_bronze;
```

####Crear tablas para archivos CSV

```sql

CREATE TABLE IF NOT EXISTS movie_bronze.movies (
  movieId INT,
  title STRING,
  budget DOUBLE,
  homePage STRING,
  overview STRING,
  yearReleaseDate DATE,
  revenue DOUBLE,
  durationTime INT,
  movieStatus STRING,
  voteAverage DOUBLE,
  voteCount INT
)
USING CSV
OPTIONS ('header' 'true')
LOCATION 'dbfs:/mnt/mymoviehistory/bronze/movie.csv'
```

```sql
select * from movie_bronze.movies;
```

```sql
DROP TABLE IF EXISTS movie_bronze.genre;
CREATE TABLE IF NOT EXISTS movie_bronze.genre (
  genreId INT,
  genreName STRING
)
USING CSV
OPTIONS ('header' 'true')
LOCATION 'dbfs:/mnt/mymoviehistory/bronze/genre.csv'
```

```sql
select * from movie_bronze.genre;
```

```sql
CREATE TABLE IF NOT EXISTS movie_bronze.language (
  languageId INT,
  languageCode STRING,
  languageName STRING
)
USING CSV
OPTIONS ('header' 'true')
LOCATION 'dbfs:/mnt/mymoviehistory/bronze/language.csv'
```

```sql
select * from movie_bronze.language;
```

```sql
drop table if exists movie_bronze.countries;
CREATE TABLE IF NOT EXISTS movie_bronze.countries (
  countryId INT,
  countryIsoCode STRING,
  countryName STRING
)
USING JSON
OPTIONS (path "/mnt/mymoviehistory/bronze/country.json")
```

```sql
select * from movie_bronze.countries;
```

```sql
drop table if exists movie_bronze.persons;
CREATE TABLE IF NOT EXISTS movie_bronze.persons (
  personId INT,
  personName STRUCT<forename: STRING, surname: STRING>
)
USING JSON
OPTIONS (path "/mnt/mymoviehistory/bronze/person.json")
```

```sql
select * from movie_bronze.persons;
```

```sql
drop table if exists movie_bronze.movie_genres;
CREATE TABLE IF NOT EXISTS movie_bronze.movie_genres (
  movieId INT,
  genreId INT
)
USING JSON
OPTIONS (path "/mnt/mymoviehistory/bronze/movie_genre.json")
```

```sql
select * from movie_bronze.movie_genres;
```

```sql
drop table if exists movie_bronze.movies_cats;
CREATE TABLE IF NOT EXISTS movie_bronze.movies_cats (
  movieId INT,
  personId STRING,
  characterName STRING,
  genderId INT,
  castOrder INT
)
USING JSON
OPTIONS (path "/mnt/mymoviehistory/bronze/movie_cast.json", multiline 'true')
```

```sql
select * from movie_bronze.movies_cats;
```

```sql
drop table if exists movie_bronze.languages_roles;
CREATE TABLE IF NOT EXISTS movie_bronze.languages_roles (
  roleId INT,
  languageRole STRING
)
USING JSON
OPTIONS (path "/mnt/mymoviehistory/bronze/language_role.json", multiline 'true')
```

```sql
select * from movie_bronze.languages_roles;
```
