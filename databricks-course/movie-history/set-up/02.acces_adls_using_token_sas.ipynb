**Acceder a Azure Data Lake Storage mediante Token SAS**
- 1.- Establecer la configuraci√≥n de spark "SAS Token"
- 2.- Listar archivos del contenerdor "demo"
- 3.- Leer del archivo "movie.csv" 

```python
movie_sas_token=dbutils.secrets.get(scope="movie-history-secret-scope",key="movie-sas-token")
```

```python
spark.conf.set("fs.azure.account.auth.type.mymoviehistory.dfs.core.windows.net", "SAS")
spark.conf.set("fs.azure.sas.token.provider.type.mymoviehistory.dfs.core.windows.net", 
               "org.apache.hadoop.fs.azurebfs.sas.FixedSASTokenProvider")
spark.conf.set("fs.azure.sas.fixed.token.mymoviehistory.dfs.core.windows.net", 
               movie_sas_token)
```

```python
display(dbutils.fs.ls("abfss://demo@mymoviehistory.dfs.core.windows.net"))
```

```python
display(spark.read.csv("abfss://demo@mymoviehistory.dfs.core.windows.net/movie.csv"))
```
