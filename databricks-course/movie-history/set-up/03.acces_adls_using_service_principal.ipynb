**Acceder a Azure Data Lake Storage mediante Service Principal**
- 1.- Registrar la aplicación en Azure Entra ID /  Service Principal
- 2.- Generar un secreto (contraseña) para la aplicación
- 3.- Configurar Spark con APP / Client ID, Directory / Tenand ID & Secret
- 4.- Asignar el Role Storage Blob Data Contributor al Data Lake

```python
client_id=dbutils.secrets.get(scope="movie-history-secret-scope",key="client-id")
directory_id=dbutils.secrets.get(scope="movie-history-secret-scope",key="directory-id")
client_secret=dbutils.secrets.get(scope="movie-history-secret-scope",key="client-secret")
```

```python
spark.conf.set("fs.azure.account.auth.type.mymoviehistory.dfs.core.windows.net", "OAuth")
spark.conf.set("fs.azure.account.oauth.provider.type.mymoviehistory.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set("fs.azure.account.oauth2.client.id.mymoviehistory.dfs.core.windows.net", client_id)
spark.conf.set("fs.azure.account.oauth2.client.secret.mymoviehistory.dfs.core.windows.net", client_secret)
spark.conf.set("fs.azure.account.oauth2.client.endpoint.mymoviehistory.dfs.core.windows.net", f"https://login.microsoftonline.com/{directory_id}/oauth2/token")
```

```python
display(dbutils.fs.ls("abfss://demo@mymoviehistory.dfs.core.windows.net"))
```
