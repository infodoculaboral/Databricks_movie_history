**Acceder a Azure Data Lake Storage mediante Access Key**
- 1.- Establecer la configuraci√≥n de spark "fs.azure.account.key"
- 2.- Listar archivos del contenerdor "demo"
- 3.- Leer del archivo "movie.csv" 

```python
movie_access_key=dbutils.secrets.get(scope="movie-history-secret-scope",key="movie-access-key")
```

```python
spark.conf.set(
    "fs.azure.account.key.mymoviehistory.dfs.core.windows.net",
    movie_access_key
)
```

```python
display(dbutils.fs.ls("abfss://demo@mymoviehistory.dfs.core.windows.net"))
```

```python
display(spark.read.csv("abfss://demo@mymoviehistory.dfs.core.windows.net/movie.csv"))
```
