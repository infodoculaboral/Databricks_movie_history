###Joins

```sql
use movie_silver;
```

```sql
desc movies;
```

```sql
create or replace temp view v_movies_2011 as
select movie_id, title, budget, revenue, year_release_date, release_date 
from movies
where year_release_date = 2011;
```

```sql
select * from v_movies_2011;
```

```sql
create or replace temp view v_productions_countries as
select * 
from productions_countries;
```

```sql
select * from productions_countries;
```

```sql
create or replace temp view v_countries as
select country_id, country_name 
from countries;
```

```sql
select * from v_countries;
```

####Inner join

```sql
select vm.title, vm.budget, vm.revenue, vm.year_release_date, vc.country_name
from v_movies_2011 vm
inner join v_productions_countries vpc
on vm.movie_id = vpc.movie_id
inner join v_countries vc
on vc.country_id = vpc.country_id;
```

####Left join

```sql
select vm.title, vm.budget, vm.revenue, vm.year_release_date, vc.country_name
from v_movies_2011 vm
left join v_productions_countries vpc
on vm.movie_id = vpc.movie_id
left join v_countries vc
on vc.country_id = vpc.country_id;
```

####Right join

```sql
select vm.title, vm.budget, vm.revenue, vm.year_release_date, vc.country_name
from v_movies_2011 vm
right join v_productions_countries vpc
on vm.movie_id = vpc.movie_id
right join v_countries vc
on vc.country_id = vpc.country_id;
```

####Full join

```sql
select vm.title, vm.budget, vm.revenue, vm.year_release_date, vc.country_name
from v_movies_2011 vm
full join v_productions_countries vpc
on vm.movie_id = vpc.movie_id
full join v_countries vc
on vc.country_id = vpc.country_id;
--hace un left, un right y un inner join unidos, mostrando el resultado con los nulls correspondientes a cada caso.
```

####Semi join

```sql
select vm.title, vm.budget, vm.revenue, vm.year_release_date
from v_movies_2011 vm
inner join v_productions_countries vpc
on vm.movie_id = vpc.movie_id
semi join v_countries vc
on vc.country_id = vpc.country_id;
--el semi join sólo devuelve las columnas de la consulta de la izquierda
```

####Anti join

```sql
select vm.title, vm.budget, vm.revenue, vm.year_release_date
from v_movies_2011 vm
inner join v_productions_countries vpc
on vm.movie_id = vpc.movie_id
anti join v_countries vc
on vc.country_id = vpc.country_id;
--datos que no existen en v_countries y si en el join anterior, y no se da ningún caso
```

####Cross join

```sql
select vm.title, vm.budget, vm.revenue, vm.year_release_date, vc.country_name
from v_movies_2011 vm
cross join v_countries vc;
--producto cartesiano de ambas tablas
```
