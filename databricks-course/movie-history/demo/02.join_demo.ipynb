###Spark Join Transformation

####INNER JOIN

```python
%run "../includes/configuration"
```

```python
movie_df = spark.read.parquet(f"{silver_folder_path}/movies").filter("year_release_date = 2007")
```

```python
production_country_df = spark.read.parquet(f"{silver_folder_path}/productions_countries")
```

```python
country_df = spark.read.parquet(f"{silver_folder_path}/countries")
```

```python
display(movie_df)
```

```python
display(production_country_df)
```

```python
display(country_df)
```

```python
movie_production_country_df = movie_df \
                                .join(production_country_df, "movie_id") \
                                .join(country_df, "country_id") \
                                .select(movie_df.title,movie_df.budget,country_df.country_name)
display(movie_production_country_df)
```

#### Outer Join -- Left Outer Join

```python
movie_production_country_df = movie_df \
                                .join(production_country_df, "movie_id", "left") \
                                .join(country_df, "country_id", "left") \
                                .select(movie_df.title,movie_df.budget,country_df.country_name)
display(movie_production_country_df)
```

#### Outer Join -- Right Outer Join

```python
movie_production_country_df = movie_df \
                                .join(production_country_df, "movie_id", "right") \
                                .join(country_df, "country_id", "right") \
                                .select(movie_df.title,movie_df.budget,country_df.country_name)
display(movie_production_country_df)
```


#### Outer Join -- Full Outer Join

```python
movie_production_country_df = movie_df \
                                .join(production_country_df, "movie_id", "full") \
                                .join(country_df, "country_id", "full") \
                                .select(movie_df.title,movie_df.budget,country_df.country_name)
display(movie_production_country_df)
```

#### Outer Join -- Semi Join

```python
movie_production_country_df = movie_df \
                                .join(production_country_df, "movie_id", "left") \
                                .join(country_df, "country_id", "semi") \
                                .select(movie_df.title,movie_df.budget)
display(movie_production_country_df)
#no puedo poner title porque con el semi, pierdo las columnas de la tabla de la derecha al no ser coincidentes
```

#### Outer Join -- Anti Join

```python
movie_production_country_df = movie_df \
                                .join(production_country_df, "movie_id", "left") \
                                .join(country_df, "country_id", "anti") \
                                .select(movie_df.title,movie_df.budget)
display(movie_production_country_df)
```

#### Outer Join -- Cross Join
producto cartesiano

```python
movie_production_country_df = movie_df.crossJoin(production_country_df) 
display(movie_production_country_df)
```
